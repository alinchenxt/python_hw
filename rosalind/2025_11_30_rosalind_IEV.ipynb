## Calculating Expected Offspring (IEV)

```python
s = "19372 16304 19636 18360 16884 20000"
l = s.split(" ")

gens = {}
for i in range(0, 6):
    if i == 0:
        gens["gen_1"] = l[0]
    elif i == 1:
        gens["gen_2"] = l[1]
    elif i == 2:
        gens["gen_3"] = l[2]
    elif i == 3:
        gens["gen_4"] = l[3]
    elif i == 4:
        gens["gen_5"] = l[4]
    elif i == 5:
        gens["gen_6"] = l[5]


for key in gens:
    gens[key] = int(gens[key])


offspr = 0
for gen in gens:
    if gen == "gen_1":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((4 * gens["gen_1"])/2)
    if gen == "gen_2":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((4 * gens["gen_2"])/2)
    if gen == "gen_3":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((4 * gens["gen_3"])/2)
    if gen == "gen_4":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((3 * gens["gen_4"])/2)
    if gen == "gen_5":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((2 * gens["gen_5"])/2)
    if gen == "gen_6":
        if gens[gen] == 0: 
            pass
        else:
            offspr += ((0 * gens["gen_6"])/2)
    
print(offspr)



```
