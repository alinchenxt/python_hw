## Transcribing DNA into RNA (RNA)

```python
dna = "GATGGAACTTGACTACGTAAATT"
rna = ""

for base in list(dna):
    if base == "T": 
        rna = rna + base.replace("T", "U")
    else:
        rna = rna + base

print(rna)
```
