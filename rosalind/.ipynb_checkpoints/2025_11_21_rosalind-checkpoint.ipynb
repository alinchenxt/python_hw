## Finding A Motif (SUBS)

```python
s = "GATATATGCATATACTT"
t = "ATAT"

a = s.find(t)
xx = [a+1]

for i, motif in enumerate(s):
    if a != -1:
        xx.append((s.find(t, a+1, )) + 1)
        a = (s.find(t, a+1, ))
    elif a == -1:
        pass

del xx[-1]

yy = map(str, xx)

b = " ".join(yy)
print(b)
```

## Rabbits and Recurrence Relations (FIB)

```python
#if k = 2
#n1 = 1
#n2 = 1
#n3 = n2 + 2*n1 = 3 
#n4 = n3 + 2*n2 = 5
#n5 = n4 + 2*n3 = 11

n = 28
k = 5
i = 1
d = 1

for l in range(2, n):
    if i == 1:   
        i = (i) + (k*d)
        d = k+1
        a = 1
    else:
        i = (i) + (k*a)
        a = d
        d = i
print(i)
```
