## Complementing a Strand of DNA

```python
#reverse --> replace

dna = "AAAACCCGGT"
reverse = ""

for base in dna:
    reverse = base + reverse

replace = ""

for nucleo in reverse:
    if nucleo == "A":
        replace = replace + "T"
    if nucleo == "C":
        replace = replace + "G"
    if nucleo == "T":
        replace = replace + "A"
    if nucleo == "G":
        replace = replace + "C"

print(replace)
```
