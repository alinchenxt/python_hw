```python
input = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3"""

input_lines = input.split('\n')

input_split = []
for item in input_lines:
    input_split.append(item.split(','))


rectangles = set()
for item in input_split:
    position = int(item[0])
    line = int(item[1])
    for i, number in enumerate(input_split):
        item = input_split[i]
        next_position = int(item[0])
        x = position - next_position
        if x < 0:
            x = x * (-1)
        else: pass
        y = x + 1
        next_line = int(item[1])
        z = line - next_line
        if z < 0:
            z = z * (-1)
        a = z + 1
        b = y * a
        rectangles.add(b)

print(rectangles)

result = max(rectangles)
print(result)
```
