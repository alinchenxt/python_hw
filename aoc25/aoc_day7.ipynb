```python
input = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

input_split = input.split("\n")

len_line = len(input_split[3])

input_process = []
for item in input_split:
    if "^" in item:
        input_process.append(item)

input_separated = []
for item in input_process:
    curr_item = item
    curr_input = []
    for i in range(0, len_line):
        curr_input.append(curr_item[i])
    input_separated.append(curr_input)

########################################################

splitters = []
for i, item in enumerate(input_separated):
    curr_item = input_separated[i]
    line = i
    pos = []
    for i in range(0, len_line):
        if curr_item[i] == '^': 
            pos.append(i)
    splitters.append([line, pos])

splitters_count = 0
for item in splitters:
    splitters_count += 1

########################################################

count = 0
for i in range(0, (splitters_count - 1)):
    curr_item = splitters[i]
    next_item = splitters[i + 1]
    curr_to_check = curr_item[1]
    next_to_check = next_item[1]
    for i, item in enumerate(curr_to_check):
        curr_num = curr_to_check[i]
        if (curr_num + 1) in next_to_check or (curr_num - 1) in next_to_check:
            count += 1

last_line = splitters[splitters_count - 1]
for item in last_line[1]:
    count += 1

print(count)
```
