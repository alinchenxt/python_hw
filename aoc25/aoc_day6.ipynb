```python
lines = []
with open("/Users/alina/Desktop/python_hw/data/aoc_6_data.txt", 'r') as infile:
    for line in infile.readlines():
        cleaned = line.rstrip()
        lines.append(cleaned)

input_lines = lines

len_line = len(input_lines[0])

###################################################

def line_as_variable(input_lines, i):
    curr_line = input_lines[i]
    return curr_line

line_one = line_as_variable(input_lines, 0)
line_two = line_as_variable(input_lines, 1)
line_three = line_as_variable(input_lines, 2)
line_four = line_as_variable(input_lines, 3)
line_five = line_as_variable(input_lines, 4)

####################################################

position = []
curr = 0
for i in range(0, len_line):
    if line_one[i] == line_two[i] == line_three[i] == line_four[i] == line_five[i] == " ":
        position.append([line_one[curr:i] + "," + line_two[curr:i] + "," + line_three[curr:i] + "," + line_four[curr:i]])
        curr = (i + 1)

position.append([line_one[curr:] + "," + line_two[curr:] + "," + line_three[curr:] + "," + line_four[curr:]])

############################################################

line_five_cleaned = []
line_five_split = line_five.split(" ")
for item in line_five_split:
    if item != "":
        line_five_cleaned.append(item)

###########################################################

def four_positions(item_split):
    items_four = []
    for item in item_split:
        if len(item) == 1:
            item_four = "000" + item
            items_four.append(item_four)
        elif len(item) == 2:
            item_four = "00" + item
            items_four.append(item_four)
        elif len(item) == 3:
            item_four = "0" + item
            items_four.append(item_four)
        elif len(item) == 4:
            items_four.append(item)
    return items_four

###########################################################

position_new = []
for i, item in enumerate(position):
    curr_item = str(position[i])
    curr_item_cleaned = curr_item[2:-2]
    item_split = curr_item_cleaned.split(",")
    position_new.append(four_positions(item_split))

position_cleaned = []
for item in position_new:
    if item == []: pass
    else: position_cleaned.append(item)

##########################################################

def count_pos(curr_item, i):
    curr_item = item[i]
    return curr_item

def cat(number):
    output = ""
    for item in number:
        output += item
    return output

###########################################################

position_fin = []
for item in position_cleaned:
    curr_item = item
    one = []
    two = []
    three = []
    four = []
    position_item = []
    for item in curr_item:
        one.append(count_pos(item, 0))
        two.append(count_pos(item, 1))
        three.append(count_pos(item, 2))
        four.append(count_pos(item, 3))
    position_item.append(cat(one))
    position_item.append(cat(two))
    position_item.append(cat(three))
    position_item.append(cat(four))
    position_fin.append(position_item)

#########################################################

calculation = []
for i, item in enumerate(position_fin):
    curr_item = position_fin[i]
    if line_five_cleaned[i] == "+":
        addition = 0
        for item in curr_item:
            addition += int(item)
        calculation.append(addition)
    elif line_five_cleaned[i] == "*":
        multiplication = 1
        for item in curr_item:
            if multiplication * int(item) == 0:
                pass
            else:
                multiplication *= int(item)
        calculation.append(multiplication)

###########################################################

fin = 0
for item in calculation:
    fin += int(item)

print(fin)
```
