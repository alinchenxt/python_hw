```python
input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """

input_lines = input.split("\n")

##################################################

all_lines = []
for item in input_lines:
    line_cleaned = []
    line = item
    line_split = line.split(" ")
    for item in line_split:
        if item != "":
            line_cleaned.append(item)
    all_lines.append(line_cleaned)

####################################################

line_len = 0
for item in line_cleaned:
    line_len += 1

#####################################################
    
def counting_position(all_lines, i):
    positions = []
    for line in all_lines:
        curr_line = line
        positions.append(line[i])
    return positions

#####################################################

sum = []
for i in range(0, line_len):
    calculation = (counting_position(all_lines, i))
    if calculation[-1] == "+":
        addition = 0
        del calculation[-1]
        for item in calculation:
            addition += int(item)
        sum.append(addition)
    elif calculation[-1] == "*":
        multiplication = 1
        del calculation[-1]
        for item in calculation:
            multiplication *= int(item)
        sum.append(multiplication)

###################################################

final = 0 
for item in sum:
    final += int(item)

print(final)
```
