```python
input = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]

def max_number(position, item):
    len_item = len(item)
    max_item = 0
    for i in range(0, len_item - position):
        if int(item[i]) > max_item:
            max_item = int(item[i])
            max_pos = i
    return max_pos

def twelve_item(item):
    position = 11
    twelve = []
    for i in range(0, 12):
         max_pos = max_number(position, item)
         twelve.append(item[max_pos])
         item = item[(max_pos + 1):]
         if i < 11:
             position = 10 - i
    twelve_cat = ""
    for item in twelve:
        twelve_cat += item
    return twelve_cat

final = []
for item in input:
    final.append(twelve_item(item))

result = 0
for item in final:
    result += int(item)

print(result)
```
