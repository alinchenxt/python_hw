```python
input = """aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out"""


input_split = input.split("\n")


lines = []
for item in input_split:
    curr_item = []
    item_split = item.split(" ")
    for trio in item_split: curr_item.append(trio)
    lines.append(curr_item)


for item in lines:
    if item[0] == "you:": start = item


start = start[1:]

#############################################

check = start
result = 0
while check != []:
    curr = check.pop(0)
    curr_str = str(curr)
    for item in lines:
        if curr_str in item[0]:
            if 'out' in item: result += 1
            else: 
                for trio in item:
                    if ":" not in trio: check.append(trio)

print(result)
```
