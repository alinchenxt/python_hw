```python
def each_input_position(input):
    input_positions = []
    for item in input:
        curr_item = str(item)
        positions_per_item = []
        for i, item in enumerate(curr_item):
            if curr_item[i] == "\'":
                pass
            else:
                positions_per_item.append(curr_item[i])
        input_positions.append(positions_per_item)
    return input_positions


def scrolls_first_line(input, max_len):
    scrolls_count = []
    curr_line0 = input[0]
    next_line0 = input[1]
    for i in range (0, max_len):
        scrolls = []
        if curr_line0[i] == "@":
            if i != 0:
                coor_seven0 = next_line0[i-1]
                scrolls.append(coor_seven0)
                coor_eight0 = curr_line0[i-1]
                scrolls.append(coor_eight0)
            else: pass
            coor_six0 = next_line0[i]
            scrolls.append(coor_six0)
            if i != (max_len -1):
                coor_four0 = curr_line0[i+1]
                scrolls.append(coor_four0)
                coor_five0 = next_line0[i+1]
                scrolls.append(coor_five0)
            else: pass
            scrolls_per_coor = 0
            for item in scrolls:
                if item == "@":
                    scrolls_per_coor +=1
            scrolls_count.append(scrolls_per_coor)
        else: pass
    return scrolls_count


def scrolls_middle_lines(input, max_len):
    scrolls_count = []
    for i in range(0, (max_len - 2)):
        prev_line = input[i]
        curr_line = input[i+1]
        next_line = input[i+2]
        for i in range (0, max_len):
            scrolls = []
            if curr_line[i] == "@":
                if i != 0:
                    coor_one = prev_line[i-1]
                    scrolls.append(coor_one)
                    coor_seven = next_line[i-1]
                    scrolls.append(coor_seven)
                    coor_eight = curr_line[i-1]
                    scrolls.append(coor_eight)
                else: pass
                coor_two = prev_line[i]
                scrolls.append(coor_two)
                coor_six = next_line[i]
                scrolls.append(coor_six)
                if i != (max_len -1):
                    coor_three = prev_line[i+1]
                    scrolls.append(coor_three)
                    coor_four = curr_line[i+1]
                    scrolls.append(coor_four)
                    coor_five = next_line[i+1]
                    scrolls.append(coor_five)
                else: pass
                scrolls_per_coor = 0
                for item in scrolls:
                    if item == "@":
                        scrolls_per_coor +=1
                scrolls_count.append(scrolls_per_coor)
            else: pass
    return scrolls_count

def scrolls_per_middle_lines(input, max_len):
    scrolls_count = []
    for i in range(0, (max_len - 2)):
        prev_line = input[i]
        curr_line = input[i+1]
        next_line = input[i+2]
        line_count = []
        for i in range (0, max_len):
            scrolls = []
            if curr_line[i] == "@":
                if i != 0:
                    coor_one = prev_line[i-1]
                    scrolls.append(coor_one)
                    coor_seven = next_line[i-1]
                    scrolls.append(coor_seven)
                    coor_eight = curr_line[i-1]
                    scrolls.append(coor_eight)
                else: pass
                coor_two = prev_line[i]
                scrolls.append(coor_two)
                coor_six = next_line[i]
                scrolls.append(coor_six)
                if i != (max_len -1):
                    coor_three = prev_line[i+1]
                    scrolls.append(coor_three)
                    coor_four = curr_line[i+1]
                    scrolls.append(coor_four)
                    coor_five = next_line[i+1]
                    scrolls.append(coor_five)
                else: pass
                scrolls_per_coor = 0
                for item in scrolls:
                    if item == "@":
                        scrolls_per_coor +=1
                line_count.append(scrolls_per_coor)
            else: pass
        scrolls_count.append(line_count)
    return scrolls_count

def scrolls_last_line(input, max_len):
    scrolls_count = []
    prev_line_last = input[max_len - 2]
    curr_line_last = input[max_len - 1]
    for i in range (0, max_len):
        scrolls = []
        if curr_line_last[i] == "@":
            if i != 0:
                coor_one_last = prev_line_last[i-1]
                scrolls.append(coor_one_last)
                coor_eight_last = curr_line_last[i-1]
                scrolls.append(coor_eight_last)
            else: pass
            coor_two_last = prev_line_last[i]
            scrolls.append(coor_two_last)
            if i != (max_len -1):
                coor_three_last = prev_line_last[i+1]
                scrolls.append(coor_three_last)
                coor_four_last = curr_line_last[i+1]
                scrolls.append(coor_four_last)
            else: pass
            scrolls_per_coor = 0
            for item in scrolls:
                if item == "@":
                    scrolls_per_coor +=1
            scrolls_count.append(scrolls_per_coor)
        else: pass
    return scrolls_count


def number_of_scrolls(scrolls_count):
    scrolls_final = 0
    for item in scrolls_count:
        if item < 4:
            scrolls_final += 1
    return scrolls_final


def all_scrolls_clean(all_scrolls):
    all_scrolls_cleaned = []
    for i, item in enumerate(all_scrolls):
        if i == 0:
            all_scrolls_cleaned.append(all_scrolls[i])
        elif i == 2:
            all_scrolls_cleaned.append(all_scrolls[i])
        else:
            curr_item = all_scrolls[i]
            for item in curr_item:
                all_scrolls_cleaned.append(item)
    return all_scrolls_cleaned


def replace_positions(scrolls_per_line_cleaned, all_scrolls_cleaned):
    scrolls_final = 0
    pos_final = []
    for i, item in enumerate(scrolls_per_line_cleaned):
        curr_scroll = scrolls_per_line_cleaned[i]
        curr_position = all_scrolls_cleaned[i]
        pos_line = []
        for i, item in enumerate(curr_scroll):
            if curr_scroll[i] < 4:
                scrolls_final += 1
                pos_line.append (curr_position[i])
        pos_final.append(pos_line)
    return scrolls_final, pos_final


def first_scrolls_pos(input, max_len):
    mid_scrolls = []
    curr_line0 = input[0]
    next_line0 = input[1]
    scrolls_pos = []
    for i in range (0, max_len):
        scrolls = []
        if curr_line0[i] == "@":
            mid_scrolls.append(i)
    return mid_scrolls

def middle_scrolls_pos(input, max_len):
    mid_scrolls = []
    for i in range(0, (max_len - 2)):
        prev_line = input[i]
        curr_line = input[i+1]
        next_line = input[i+2]
        scrolls_pos = []
        for i in range (0, max_len):
            scrolls = []
            if curr_line[i] == "@":
                scrolls_pos.append(i)
        mid_scrolls.append(scrolls_pos)
    return mid_scrolls

def last_scrolls_pos(input, max_len):
    mid_scrolls = []
    scrolls_count = []
    prev_line_last = input[max_len - 2]
    curr_line_last = input[max_len - 1]
    for i in range (0, max_len):
        scrolls = []
        if curr_line_last[i] == "@":
            mid_scrolls.append(i)
    return mid_scrolls


def scrolls_cleaned(all_scrolls, scrolls_per_line):
    all_scrolls_cleaned = []
    for i, item in enumerate(all_scrolls):
        if i == 0:
            all_scrolls_cleaned.append(all_scrolls[i])
        elif i == 2:
            all_scrolls_cleaned.append(all_scrolls[i])
        else: 
            curr_item = item
            for item in curr_item:
                all_scrolls_cleaned.append(item)
    scrolls_per_line_cleaned = []
    for i, item in enumerate(scrolls_per_line):
        if i == 0:
            scrolls_per_line_cleaned.append(scrolls_per_line[i])
        elif i == 2:
            scrolls_per_line_cleaned.append(scrolls_per_line[i])
        else: 
            curr_item = item
            for item in curr_item:
                scrolls_per_line_cleaned.append(item)
    return scrolls_per_line_cleaned

def replace_with(pos_final, input_positions):
    for i, item in enumerate(pos_final):
        if pos_final[i] == []: 
            pass
        else:
            ex_pos = pos_final[i]
            input_replace = input_positions[i]
            for i, item in enumerate(ex_pos):
                replace = ex_pos[i]
                input_replace[replace] = '.'
    return input_positions
```
