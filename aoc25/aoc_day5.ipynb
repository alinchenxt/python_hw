```python
lines = []
with open("/Users/alina/Desktop/python_hw/aoc25/day_5.txt", 'r') as infile:
    for line in infile.readlines():
        cleaned = line.rstrip()
        lines.append(str(cleaned))

for i, item in enumerate(lines):
    if lines[i] == "":
        del lines[i]

input = lines

#########################################

ingredients = []
ranges = []
for item in input:
    if "-" in item:
        ranges.append(item)
    else:
        ingredients.append(item)


all_ranges = []
for item in ranges:
    split_range = item.split("-")
    start_range = int(split_range[0])
    stop_range = int(split_range[1])
    for i in range(start_range, (stop_range + 1)):
        all_ranges.append(i)

###############################################

def ingr_in_ranges(ranges, ingredient):
    curr_ingr = ""
    for item in ranges:
        if str(item) == ingredient:
            curr_ingr = str(item)
    return curr_ingr

###############################################

ingr_count = []
for item in ingredients:
    ingr_count.append(ingr_in_ranges(all_ranges, item))


for i, item in enumerate(ingr_count):
    if ingr_count[i] == "":
        del ingr_count[i]

##########################################

result = 0
for item in ingr_count:
    result += 1

print(result)
```
