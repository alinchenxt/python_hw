```python
input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

instructions = input.split("\n")

s = 50
counting = ""

for item in instructions:
    if "L" in item:
        x = item.split("L")
        y = x[1]
        z = int(y)
        s = s - z
        if s < 0:
            s = s + 100
        j = str(s)
        counting += j + " "
    elif "R" in item:
        x = item.split("R")
        y = x[1]
        z = int(y)
        s = s + z
        if s > 99:
            s = s - 100
        j = str(s)
        counting += j + " "

print(counting)
counting.count("0")


```
