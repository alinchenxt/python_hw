## gold star

```python
input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

instructions = input.split("\n")

s = 50
counting = ""

for item in instructions:
    if "L" in item:
        x = item.split("L")
        y = x[1]
        z = int(y)
        s = s - z
        if s < 0:
            s = s + 100
        j = str(s)
        counting += j + " "
    elif "R" in item:
        x = item.split("R")
        y = x[1]
        z = int(y)
        s = s + z
        if s > 99:
            s = s - 100
        j = str(s)
        counting += j + " "

print(counting)
counting.count("0")


```

## silver star

```python
input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

instructions = input.split("\n")

curr_pos = 50
counting = " "

for item in instructions:
    steps = item[1:]
    steps_int = int(steps)
    if "L" in item:
        prev_pos = curr_pos #0
        curr_pos -= steps_int #-200
        while curr_pos < 0:
            curr_pos += 100 #x2
            if prev_pos != 0:
                counting += "0 " 
            else: prev_pos += 1
        str_curr_pos = str(curr_pos) 
        counting += str_curr_pos + " " #0
    elif "R" in item:
        curr_pos += steps_int
        while curr_pos > 99:
            curr_pos -= 100
            if curr_pos != 0:
                counting += "0 "
        str_curr_pos = str(curr_pos)
        counting += str_curr_pos + " "

counting.count(" 0")
```
