### Dictionaries

* key (string or number) --> value (string or number)
* {} to define dictionaries
* dictionaries (like lists) are mutable
* len gives number of pairs: len(dict)
* del dict['key']
* dict.keys() or dict.values() or dict.items()
* items = pairs
* keys should be unique
* nested dictionaries: nested dictionary is value of top dictionary
* dictionaries are a sort of collections

```python
grades = {'Juan': {'Maths': 2},
          'John': {'History': 1, 'Biology': 3},
          'Gianni': {'Biology': 1}}

grades['Juan']['History'] = 2
grades['Juan']['Biology'] = 1

grades['John']['Maths'] = 1

grades['Gianni']['History'] = 4
grades['Gianni']['Maths'] = 3

print(grades)
```

### Loops

* indentation: 1 tab
* collections = strings, lists, dictionaries

```python
for <loop variable> in <collection (list of values)>: 
    <body (commands)>
```

```python
for number in [123, 52, 38, 41, 50]:
    double = number * 2
    print(number, double)
```

```python
s = "the quick brown fox jumps over the lazy dog"

for character in s:
    print(character)

for word in s.split():
    print(word)
```

```python
shopping = ['bread', 'milk', 'cheese']
times = ['1', '2', '3']
shop = ['oefferl', 'hofer', 'hofer']

for i in range(3): #when length of list known
    print('Buy', shopping[i], times[i], 'times at', shop[i])

for i in range(len(shopping)): #when length of list unknown
    print('Buy', shopping[i], times[i], 'times at', shop[i])

for i, item in enumerate(shopping): #when length of list unknown
    #print(i, item)
    print('Buy', shopping[i], times[i], 'times at', shop[i])
```

```python
data = [1,2,2,5]
total = 0
cumulative = []


for number in data:
    total = total + number
    cumulative.append(total)

print(cumulative)
```

```python
message = ""
for number in range(10):
    # use a if the number is a multiple of 3, otherwise use b
    if (number % 3) == 0:
        message = message + "a"
    else:
        message = message + "b"
print(message)
```
