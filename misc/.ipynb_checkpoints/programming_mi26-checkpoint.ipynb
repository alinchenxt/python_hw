### Open Reading Frame (ORF)

```python
s = "AGCCATGTAGCTAACTCAGGTTACATGGGGATGACCCCGCGACTTGGATTAGAGTCTCTTTTGGAATAAGCCTGAATGATCCGAGTAGCATCTCAG"
u = s.replace("T", "U")
g = s[::-1]
r = g.replace("T", "U")

comp = ""
for base in r:
    if base == "A": comp += "U"
    if base == "C": comp += "G"
    if base == "G": comp += "C"
    if base == "U": comp += "A"

start = []
for i in range(0, len(comp)):
    if "AUG" == comp[i:i+3]:
        start.append(i)


orfs = []

for item in start:
    triple = ""
    for i in range(item, len(comp), 3):
        codon = comp[i:i+3]
        if codon in ("UAG", "UGA", "UAA"):
            triple = triple + " " + codon
            orfs.append(triple)
            break
        else:
            triple = triple + " " + codon


start = []
for i in range(0, len(r)):
    if "AUG" == u[i:i+3]:
        start.append(i)


for item in start:
    triple = ""
    for i in range(item, len(u), 3):
        codon = u[i:i+3]
        if codon in ("UAG", "UGA", "UAA"):
            triple = triple + " " + codon
            orfs.append(triple)
            break
        else:
            triple = triple + " " + codon

print(orfs)

prot = []
x = ""

for orf in orfs:
    a = orf
    for key in a.split(" "):
        x = x + codon_table[key]
        prot.append(x)

print(prot)

#for key in p.split(" "):
 #   x = x + codon_table[key]

#print(x)

#orfs_uniq = []
#for orf in orfs:
#    if orf not in orfs_uniq:
#        orfs_uniq.append(orf)
#print(orfs_uniq)
```
