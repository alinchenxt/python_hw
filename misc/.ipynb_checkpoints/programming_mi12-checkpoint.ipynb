### Conditionals (if statements)

if <condition>:
    <something>
else:
    <do_something_else>

* elif = else if
* conditionals
    * test equality with == (number == 2)
    * test inequality with != (number != 2)

```python
number = 4
if number > 0:
    print("this is a positive number")
else:
    print("this is a negative number")
```

```python
number = 0
if number > 0:
    print("this is a positive number")
elif number < 0:
    print("this is a negative number")
else:
    print("this number is 0")
```

```python
fruit = "apple"
if fruit == "banana":
    print("this is a banana")
else:
    print("this is a(n)", fruit)
```

```python
fruits = ["apple", "banana", "grape", "orange", "pomelo", "ananas", "kiwi", "banana"]

for i in range(len(fruits)):
    fruit = (fruits[i])
    print("this fruit is a(n)", fruit)
    if fruits[i] == "banana":
        print("found a banana")
```

```python
fruits = ["apple", "banana", "grape", "orange", "pomelo", "ananas", "kiwi", "banana"]

for i in range(len(fruits)):
    if fruits[i] == "banana":
        print("found a banana at position", i)
```

```python
abc = "We tried list and we tried dicts also we tried Zen"
abcd = abc.split()

dict = {}

for word in abcd:
    dict[word] = abcd.count(word)

for key, value in dict.items():
    print(key, value)
```

```python
abc = "We tried list and we tried dicts also we tried Zen"
words = abc.split
dict = {}

for word in words:
    dict[word] = 0

for word in words:
    dict[word] = dict[word] + 1

for key, value in dict.items():
    print(key, value)
```
